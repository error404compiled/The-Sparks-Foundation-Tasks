### US Terror Attacks and Death Injuries.

#### nkillus= No. of fatalities
#### nwoundus= No. of non fatalities

```python
df.nkillus.plot(kind = 'line', color = 'red', label = 'The Number of Total Confirmed Fatalities', figsize = (10,10), grid = True, 
                linestyle = ':', fontsize=10)

df.nwoundus.plot(color = "blue", label = 'The Number of Confirmed Non-Fatal Injuries',figsize = (10,10), grid = True,
                 linestyle = '-.',  fontsize=10)

# legend = puts label in upper right corner
plt.legend(loc='upper right')
plt.xlabel('Database Index', fontsize=10)
plt.ylabel('Number of Dead or Injuries', fontsize=10)

plt.title('Confirmed Fatalities & Non-Fatal Injuries')
plt.show()
```

### Death vs Injuries.

```python
df.plot(kind = 'scatter', x = 'nkill', y = 'nwound', alpha = 0.5, figsize = (10,10), fontsize=15, color = 'purple')
plt.xlabel('Kill', fontsize=15)
plt.ylabel('Wound', fontsize=15)
plt.title('Kill - Wound Scatter Plot')
plt.show()
```

### Observation:
>The mortality rate and injuries were low, but a small number of actions led to too many deaths and injuries.

### Frequency of terrorism actions by years

```python
df.iyear.plot(kind = 'hist', bins = 20, figsize = (15,10), color = 'red', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Frequency of Terrorism Actions by Years')
plt.show()
```

### Observation:
> Terrorist attacks have increased in recent years.

```python
killData = df.loc[:,'nkill']

#Dropping NaN values using dropna
print('Number of people killed by terror attack:', int(sum(killData.dropna())))
```

### Let's look at what types of attacks these deaths were made of.

```python
attackData = df.loc[:,'attacktype1':'attacktype1_txt']
#attacktype1 is the attack id and attacktype1_txt is the attack type
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
labels = typeKillFormatData.columns.tolist() # convert line to list
transpoze = typeKillFormatData.T # transpoze
values = transpoze["nkill"]

fig, ax = plt.subplots(figsize=(17, 17), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, fontsize=10,loc='upper right') # location legend

plt.show()
```

### Observation:
>Armed assault and bombing/explosion are seen to be the cause of 77% of the deaths in these attacks.
